```python
rating_with_name = rating.merge(book,on='ISBN')
```

```python
num_rating_df = rating_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns={'Book-Rating':'Num_rating'},inplace=True)
num_rating_df
```

```python
avg_rating_df = rating_with_name.groupby('Book-Title').mean(numeric_only=True)['Book-Rating'].reset_index()
avg_rating_df.rename(columns={'Book-Rating':'Avg_rating'},inplace=True)
avg_rating_df
```

```python
popular_df = num_rating_df.merge(avg_rating_df,on='Book-Title')
popular_df
```

```python
pbr_df=popular_df[popular_df['Num_rating']>=300].sort_values('Avg_rating',ascending=False).head(50)
# pbr_df = popular_df.sort_values('Avg_rating', ascending=False).head(250)
```

```python
pbr_df= pbr_df.merge(book,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Publisher','Image-URL-M','Num_rating','Avg_rating']]
```

```python
b = rating_with_name.groupby('User-ID').count()['Book-Rating']>250
users_with_ratings = b[b].index
```

```python
filtered_rating = rating_with_name[rating_with_name['User-ID'].isin(users_with_ratings)]
```

```python
c = filtered_rating.groupby('Book-Title').count()['Book-Rating'] >= 50
famous_books = c[c].index
```

```python
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
pt = final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(pt)
```

```python
similarity_scores.shape
```

```python
def recommendation(book_name):
    # fetching Index
    index = np.where(np.array(list(pt.index))==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[0])),reverse=True,key=lambda x:x[1])[1:9]
    
    data =[]
    for i in similar_items:
        item = []
        temp_df = book[book['Book-Title'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        
        data.append(item)
    return data
```

```python
recommendation("Harry Potter and the Sorcerer's Stone (Harry Potter (Paperback))")
```

```python
def recommendation_categories(categories_name=None):
    if categories_name is None:
        return "Please provide a category name."
    
    filtered_books = dt[dt['categories'] == categories_name][['title', 'average_rating']]
    sorted_books = filtered_books.sort_values(by='average_rating', ascending=False)
    
    if len(sorted_books) >= 10:
        return sorted_books.head(10)
    else:
        return sorted_books
```
